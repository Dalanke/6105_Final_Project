```python
# Need to remove NA
# Regression - Try to predict PM2.5 (y) from non PM2.5 columns (X)
# parameter: data, n(rows to shift), dependent_col, drop_cols(including dependent_col)
def valid(col,drop_cols):
    return col not in drop_cols

def randomforest(data,n,dependent_col,drop_cols):
    X = data[[x for x in data.columns if valid(x,drop_cols)]]
    y = data[dependent_col]
    if n>0:
        X = X.iloc[:-n]
        y = y.shift(-n).dropna()
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
    # Create a model 
    rf_model = RandomForestRegressor(oob_score=True)
    # Train the model
    rf_model.fit(X_train, y_train)
    
    return rf_model,X_test, y_test
```

# 1.5 Combine six hours data

```python
data_shunyi
```

```python
#drop cols that don't need to average
data_6hours_combination = data_shunyi.drop(['No','station'],axis=1)
data_6hours_combination[:6]
```

```python
a = data_6hours_combination.ix[2:2]
a.mean()
a.mean().to_frame().T
```

```python
def data_transfer(data):
    length = len(data)
    a = data.ix[:5]
    #print(type(data_test.ix[:6]))
    new_data = a.mean().to_frame().T
    index=6
    while(index+5<len(data)):
        hours_data = data.ix[index:5+index]
        #print(hours_data)
        tmp_data = hours_data.mean().to_frame().T
        new_data = pd.concat([new_data,tmp_data],axis=0)
        index=index+6
    return new_data

data_6hours_combination = data_transfer(data_6hours_combination)
data_6hours_combination.head()
```

```python
num_null = data_6hours_combination[data_6hours_combination.isnull().T.any()]
num_null
```

```python
data_6hours_combination.dropna(inplace=True)
data_6hours_combination
```

```python
#recall random forest
drop_cols_6hours = ['year','day','hour','PM2.5','PM10']
dependent_col_6hours = 'PM2.5'
rf_model_6hours,X_test_6hours, y_test_6hours = randomforest(data_6hours_combination,1,dependent_col_6hours,drop_cols_6hours)
print("accurate:", rf_model_6hours.score(X_test_6hours, y_test_6hours))
print()
print("predicted data and real data:")
print(pd.concat([pd.Series(rf_model_6hours.predict(X_test_6hours)), y_test_6hours.reset_index(
drop=True)], axis=1))
print()
print("feature importances:")
print(sorted(zip(X_test_6hours.columns, rf_model_6hours.feature_importances_),
    key=lambda x: x[1], reverse=True))
```

# 1.6 Normalization (Concentration)

```python
data_copy_concentration = data_concentration.copy()
data_copy_concentration
```

```python
data_prenormalized = data_concentration.drop(['No','year','month','day','hour','PM10','station'],axis = 1)
data_prenormalized.columns
```

```python
data_wd = data_prenormalized[[col for col in data_prenormalized if col.startswith('wd_')]]
data_wd
```

```python
data_without_wd = data_prenormalized.drop(data_wd.columns,axis=1)
data_without_wd
```

```python
data_normalized_without_wd = data_without_wd.apply(lambda x: (x - np.min(x)) / (np.max(x) - np.min(x)))
data_normalized_without_wd
```

```python
data_normalized = pd.concat([data_normalized_without_wd,data_wd],axis=1)
data_normalized
```

```python
data_normalized.columns
```

```python
#recall random forest
#drop_cols = ['PM2.5','PM10','station']
dependent_col_normalized = 'PM2.5'
rf_model_normalized,X_test_normalized, y_test_normalized = randomforest(data_normalized,1,dependent_col_normalized,dependent_col_normalized)
print("accurate:", rf_model_normalized.score(X_test_normalized, y_test_normalized))
print()
print("predicted data and real data:")
print(pd.concat([pd.Series(rf_model_normalized.predict(X_test_normalized)), y_test_normalized.reset_index(
drop=True)], axis=1))
print()
print("feature importances:")
print(sorted(zip(X_test_normalized.columns, rf_model_normalized.feature_importances_),
    key=lambda x: x[1], reverse=True))
```
