```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read data
data = pd.read_csv('data/cleanup/Shunyi.csv', index_col=['date'])
data
```

```python
data['PM2.5(t+1)'] = data['PM2.5'].shift(periods=-1)
data
```

```python
data.dropna(inplace=True)
data
```

```python
import seaborn
data_corr = data.corr()
plt.figure(figsize=(15,15))
seaborn.heatmap(data_corr, center=0, annot=True)
```

```python
# x1:PM10
X1 = data['PM10'].to_numpy()
# x2:NO2
X2 = data['NO2'].to_numpy()
# x3:CO
X3 = data['CO'].to_numpy()

# observation
Y = data['PM2.5(t+1)'].to_numpy()
```

```python
import pymc3 as pm

with pm.Model() as pm25_model_1:
    alpha1 = pm.Normal('alpha1', mu=0, sd=10)
    alpha2 = pm.Normal('alpha2', mu=0, sd=10)
    alpha3 = pm.Normal('alpha3', mu=0, sd=10)
    c = pm.Normal('c', mu=0, sd=10)
    
    sigma = pm.HalfCauchy('sigma', beta=30)
    
    mu = alpha1 * X1 + alpha2 * X2 + alpha3 * X3 + c
    
    #'Y_obs', mu=mu, sd=sigma, observed=Y
    Y_obs = pm.Normal('Y_obs', mu=mu, sd=sigma, observed=Y)
```

```python
with pm25_model_1:
    trace = pm.sample(3000, tune=1500, target_accept=0.9)
```

```python
with pm25_model_1:
    posterior_predictive = pm.sample_posterior_predictive(trace)
    prior = pm.sample_prior_predictive()
```

```python
import arviz as az

with pm25_model_1:
    p = az.from_pymc3(trace=trace, prior=prior, posterior_predictive=posterior_predictive)
    
az.plot_posterior(p, round_to=2, credible_interval=0.95)
```

```python
# model1: PM10, NO2, CO
def model1 (X1, X2, X3):
    return 0.6 * X1 + 0.13 * X2 + 0.019 * X3 - 8.3

```

```python
data['prediction'] = data.apply(lambda x: model1(x.PM10, x.NO2, x.CO), axis=1)
data
```

```python
data.prediction.plot()
data['PM2.5(t+1)'].plot(figsize=(12,6))
```

```python
error1 = sum([(data['PM2.5(t+1)'][i]-data['prediction'][i])**2 for i in range(0,data.shape[0])])
error1
```

```python
data['error'] = data.apply(lambda x: abs(x['PM2.5(t+1)']-x['prediction'])/x['PM2.5(t+1)'], axis=1)
data[data['error'] < 0.25].shape[0]
```

```python
data['error'].describe()
```

## Model 2-try studentT

```python
# model 2: change to student T

# x1:PM10 0.9
X1 = data['PM10'].to_numpy()
# x2:CO 0.78 
X2 = data['CO'].to_numpy()
# x3:NO2 0.64
X3 = data['NO2'].to_numpy()

# observation
Y = data['PM2.5(t+1)'].to_numpy()


import pymc3 as pm

with pm.Model() as pm25_model_2:
    alpha1 = pm.Normal('alpha1', mu=0, sd=10)
    alpha2 = pm.Normal('alpha2', mu=0, sd=10)
    alpha3 = pm.Normal('alpha3', mu=0, sd=10)
    c = pm.Normal('c', mu=0, sd=10)
    
    sigma = pm.HalfCauchy('sigma', beta=30)
    nu = pm.Uniform('nu', 0, 10)
    
    mu = alpha1 * X1 + alpha2 * X2 + alpha3 * X3 + c
    
    #'Y_obs', mu=mu, sd=sigma, observed=Y
    Y_obs = pm.StudentT('Y_obs', nu, mu=mu, sd=sigma, observed=Y)
```

```python
with pm25_model_2:
    trace2 = pm.sample(3000, tune=1500, target_accept=0.9)
```

```python
with pm25_model_2:
    posterior_predictive2 = pm.sample_posterior_predictive(trace2)
    prior2 = pm.sample_prior_predictive()
```

```python
import arviz as az

with pm25_model_2:
    p2 = az.from_pymc3(trace=trace2, prior=prior2, posterior_predictive=posterior_predictive2)
    
az.plot_posterior(p2, round_to=2, credible_interval=0.95)
```

## build model for winter

```python
data
```

```python
data.index = pd.to_datetime(data.index)
```

```python
def winter(m):
    
    return [True if (month == 11 or month == 12 or month == 1 or month == 2) else False for month in m]
```

```python
data_winter = data[winter(data.index.month)]
data_winter
```

```python
data_winter['PM2.5'].plot(figsize=(16,8))
```

```python
import seaborn
data_winter_corr = data_winter.corr()
plt.figure(figsize=(15,15))
seaborn.heatmap(data_winter_corr, center=0, annot=True)
```

```python
# x1:PM10
X1 = data_winter['PM10'].to_numpy()
# x2:NO2
X2 = data_winter['NO2'].to_numpy()
# x3:CO
X3 = data_winter['CO'].to_numpy()
# x4:SO2
X4 = data_winter['SO2'].to_numpy()
# x5:O3
X5 = data_winter['O3'].to_numpy()

# observation
Y = data_winter['PM2.5'].to_numpy()
```

```python
X1.shape
```

```python
import pymc3 as pm

with pm.Model() as pm25_model_3:
    alpha1 = pm.Normal('alpha1', mu=0, sd=10)
    alpha2 = pm.Normal('alpha2', mu=0, sd=10)
    alpha3 = pm.Normal('alpha3', mu=0, sd=10)
    alpha4 = pm.Normal('alpha4', mu=0, sd=10)
    alpha5 = pm.Normal('alpha5', mu=0, sd=10)
    c = pm.Normal('c', mu=0, sd=10)
    
    sigma = pm.HalfCauchy('sigma', beta=50)
    
    mu = alpha1 * X1 + alpha2 * X2 + alpha3 * X3 + alpha4 * X4 + alpha5 * X5 + c
    
    #'Y_obs', mu=mu, sd=sigma, observed=Y
    Y_obs = pm.Normal('Y_obs', mu=mu, sd=sigma, observed=Y)
```

```python
with pm25_model_3:
    trace3 = pm.sample(3000, tune=1500, target_accept=0.9)
```

```python
with pm25_model_3:
    posterior_predictive3 = pm.sample_posterior_predictive(trace3)
    prior3 = pm.sample_prior_predictive()
```

```python
import arviz as az

with pm25_model_3:
    p3 = az.from_pymc3(trace=trace3, prior=prior3, posterior_predictive=posterior_predictive3)
    
az.plot_posterior(p3, round_to=2, credible_interval=0.95)
```

```python
# model1: PM10, NO2, CO, SO2, O3
def model2 (X1, X2, X3, X4, X5):
    return 0.64 * X1 + 0.1 * X2 + 0.018 * X3 + 0.084 * X4 + 0.098 * X5 - 17
```

```python
data_winter['prediction'] = data.apply(lambda x: model2(x.PM10, x.NO2, x.CO, x.SO2, x.O3), axis=1)
data_winter
```

```python
data_winter['prediction'].plot()
data_winter['PM2.5'].plot(figsize=(16,8))
```

```python
data_winter['error'] = data_winter.apply(lambda x: abs(x['PM2.5']-x['prediction'])/x['PM2.5'], axis=1)
data_winter[data['error'] < 0.25].shape[0]
```

```python
data_winter['error'].describe()
```
