```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read data
data = pd.read_csv('data/cleanup/Shunyi.csv', index_col=['date'])
data
```

```python
# drop 'wd' first
data.drop(columns=['wd'], inplace=True)
data
```

```python
import seaborn

# plot pair wise data corration (pearson's method)
data_corr = data.corr()
plt.figure(figsize=(15,15))
seaborn.heatmap(data_corr, center=0, annot=True)
```

```python
# add 1 hour change rate into data

data_hourly = data.copy()
data_hourly['PM1hour'] = data_hourly['PM2.5'].shift(periods=-1)
data_hourly['PM2.5Change'] = data_hourly.apply(lambda x: x.PM1hour - x['PM2.5'] if ((not pd.isna(x.PM1hour)) and (not pd.isna(x['PM2.5']))) else np.nan, axis=1)
data_hourly.dropna(inplace=True)
data_hourly.drop(columns=['PM2.5'], inplace=True)
data_hourly
```

```python
# plot pair wise data corration with new data
data_hourly_corr = data_hourly.corr()
plt.figure(figsize=(15,15))
seaborn.heatmap(data_hourly_corr, center=0, annot=True)
```

```python
import math
# convert temp and dewp into RH
data_RH = data.copy()
data_RH['RH'] = data_RH.apply(lambda x: math.exp(17.27*x.DEWP/(237.7+x.DEWP)-17.27*x.TEMP/(237.7+x.TEMP)), axis=1)
data_RH.drop(columns=['TEMP', 'DEWP'], inplace=True)
data_RH
```

```python
# exam the corr
data_RH_corr = data_RH.corr()
plt.figure(figsize=(15,15))
seaborn.heatmap(data_RH_corr, center=0, annot=True)
```

```python

# see data first
data.dropna(inplace=True)
data
```

```python
# build model that ONLY use variables that correlation coefficient > 0.5


# x1:PM10
X1 = data['PM10'].to_numpy()
# x2:NO2
X2 = data['NO2'].to_numpy()
# x3:CO
X3 = data['CO'].to_numpy()

# observation
Y = data['PM2.5'].to_numpy()
```

```python
X1.shape == X2.shape == X3.shape == Y.shape
```

```python
import pymc3 as pm

with pm.Model() as pm25_model_1:
    alpha1 = pm.Normal('alpha1', mu=0, sd=10)
    alpha2 = pm.Normal('alpha2', mu=0, sd=10)
    alpha3 = pm.Normal('alpha3', mu=0, sd=10)
    c = pm.Normal('c', mu=0, sd=10)
    
    sigma = pm.HalfCauchy('sigma', beta=30)
    
    mu = alpha1 * X1 + alpha2 * X2 + alpha3 * X3 + c
    
    #'Y_obs', mu=mu, sd=sigma, observed=Y
    Y_obs = pm.Normal('Y_obs', mu=mu, sd=sigma, observed=Y)
```

```python
with pm25_model_1:
    trace = pm.sample(3000, tune=1500, target_accept=0.9)
```

```python
with pm25_model_1:
    posterior_predictive = pm.sample_posterior_predictive(trace)
    prior = pm.sample_prior_predictive()
```

```python
import arviz as az

with pm25_model_1:
    p = az.from_pymc3(trace=trace, prior=prior, posterior_predictive=posterior_predictive)
    
az.plot_posterior(p, round_to=2, credible_interval=0.95)
```

```python
# model1: PM10, NO2, CO
def model1 (X1, X2, X3):
    return 0.62 * X1 + 0.0063 * X2 + 0.021 * X3 - 7.7
```

```python
data['prediction'] = data.apply(lambda x: model1(x.PM10, x.NO2, x.CO), axis=1)
data
```

```python
data.prediction.plot()
data['PM2.5'].plot(figsize=(12,6))
```

```python
error1 = sum([(data['PM2.5'][i]-data['prediction'][i])**2 for i in range(0,30587)])
error1
```

```python
data['error'] = data.apply(lambda x: abs(x['PM2.5']-x['prediction'])/x['PM2.5'], axis=1)
data[data['error'] < 0.25].shape[0]
```

```python
data['error'].describe()
```

## Model 2

```python
# model 2: change to student T

# x1:PM10 0.9
X1 = data['PM10'].to_numpy()
# x2:CO 0.78 
X2 = data['CO'].to_numpy()
# x3:NO2 0.64
X3 = data['NO2'].to_numpy()

# observation
Y = data['PM2.5'].to_numpy()


import pymc3 as pm

with pm.Model() as pm25_model_2:
    alpha1 = pm.Normal('alpha1', mu=0, sd=10)
    alpha2 = pm.Normal('alpha2', mu=0, sd=10)
    alpha3 = pm.Normal('alpha3', mu=0, sd=10)
    c = pm.Normal('c', mu=0, sd=10)
    
    sigma = pm.HalfCauchy('sigma', beta=20)
    nu = pm.Uniform('nu', 0, 10)
    
    mu = alpha1 * X1 + alpha2 * X2 + alpha3 * X3 + c
    
    #'Y_obs', mu=mu, sd=sigma, observed=Y
    Y_obs = pm.StudentT('Y_obs', nu, mu=mu, sd=sigma, observed=Y)
```

```python
with pm25_model_2:
    trace2 = pm.sample(3000, tune=1500, target_accept=0.9)
```

```python
with pm25_model_2:
    posterior_predictive2 = pm.sample_posterior_predictive(trace2)
    prior2 = pm.sample_prior_predictive()
```

```python
import arviz as az

with pm25_model_2:
    p2 = az.from_pymc3(trace=trace2, prior=prior2, posterior_predictive=posterior_predictive2)
    
az.plot_posterior(p2, round_to=2, credible_interval=0.95)
```

```python
# model2: PM10, CO, NO2
def model2 (X1, X2, X3):
    return 0.85 * X1 + 0.013 * X2 - 0.23 * X3 - 3.7
```

```python
data['prediction'] = data.apply(lambda x: model2(x.PM10, x.CO, x.NO2), axis=1)
data
```

```python
data.prediction.plot()
data['PM2.5'].plot(figsize=(12,6))
```
